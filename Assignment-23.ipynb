Question-1:

Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

Input  : nlnll
Output : 2


```python
def calculate_depth(preorder_str):
    def calculate_depth_recursive(preorder_str, index):
        if index >= len(preorder_str):
            return 0
        
        if preorder_str[index] == 'l':
            return 0
        
        left_child_index = index + 1
        right_child_index = calculate_depth_recursive(preorder_str, left_child_index) + 1
        return max(left_child_index, right_child_index)

    return calculate_depth_recursive(preorder_str, 0)

# Test example
preorder_input = 'nlnll'
print(calculate_depth(preorder_input))
```

 Question-2:

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

***Output:** 4 5 3 6*


```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def left_view(root):
    if root is None:
        return
    
    queue = [root]
    while queue:
        level_size = len(queue)
        for i in range(level_size):
            node = queue.pop(0)
            if i == 0:
                print(node.val, end=' ')
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

root = TreeNode(4)
root.left = TreeNode(5)
root.right = TreeNode(2)
root.right.left = TreeNode(3)
root.right.right = TreeNode(1)
root.right.left.left = TreeNode(6)
root.right.left.right = TreeNode(7)

left_view(root) 
```

Question-3:

Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def right_view(root):
    if root is None:
        return

    queue = [root]
    while queue:
        level_size = len(queue)
        for i in range(level_size):
            node = queue.pop(0)
            if i == level_size - 1:
                print(node.val, end=' ')

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

# Test example 1
# Constructing the binary tree from the first example input
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
root1.left.left = TreeNode(4)
root1.left.right = TreeNode(5)
root1.right.left = TreeNode(6)
root1.right.right = TreeNode(7)
root1.right.right.right = TreeNode(8)

print("Right view of the first tree is:", end=' ')
right_view(root1)  # Output: 1 3 7 8
print()

# Test example 2
# Constructing the binary tree from the second example input
root2 = TreeNode(1)
root2.left = TreeNode(8)
root2.left.left = TreeNode(7)

print("Right view of the second tree is:", end=' ')
right_view(root2)  # Output: 1 8 7
print()
```

Question-4:

Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
from collections import defaultdict, deque

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def bottom_view(root):
    if root is None:
        return
    
    horizontal_distance_map = defaultdict(lambda: None)
    queue = deque([(root, 0)])  # Store nodes and their corresponding horizontal distance

    while queue:
        node, horizontal_distance = queue.popleft()
        horizontal_distance_map[horizontal_distance] = node.val

        if node.left:
            queue.append((node.left, horizontal_distance - 1))
        if node.right:
            queue.append((node.right, horizontal_distance + 1))

    # Print the values of the nodes from the map in sorted order of their horizontal distance
    for distance in sorted(horizontal_distance_map.keys()):
        print(horizontal_distance_map[distance], end=' ')

# Test example 1
# Constructing the binary tree from the first example input
root1 = TreeNode(20)
root1.left = TreeNode(8)
root1.right = TreeNode(22)
root1.left.left = TreeNode(5)
root1.left.right = TreeNode(3)
root1.right.right = TreeNode(25)
root1.left.right.left = TreeNode(10)
root1.left.right.right = TreeNode(14)

print("Bottom view of the first tree is:", end=' ')
bottom_view(root1)  # Output: 5 10 3 14 25
print()
```

```python
# Test example 2
# Constructing the binary tree from the second example input
root2 = TreeNode(20)
root2.left = TreeNode(8)
root2.right = TreeNode(22)
root2.left.left = TreeNode(5)
root2.left.right = TreeNode(3)
root2.right.left = TreeNode(4)
root2.right.right = TreeNode(25)
root2.left.right.left = TreeNode(10)
root2.left.right.right = TreeNode(14)

print("Bottom view of the second tree is:", end=' ')
bottom_view(root2)  # Output: 5 10 4 14 25
print
```
